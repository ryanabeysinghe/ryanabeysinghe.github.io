```python
# Import Libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm

# From Libraries
from sklearn import linear_model
from scipy import stats
```

```python
# Read the csv file on Covid 19 Data in the United States
covid_df = pd.read_csv("USA_COVID19_DATA.csv")

# Order all the States in the 'State' column in alphabetical order
covid_df = covid_df.sort_values('State', ascending = True)

# Print all the columns of the dataset
pd.set_option('display.max_columns', None)

# Print out dataset
covid_df
```

<font size = "5"> **Analyzing the Data** </font> 
***
<font size = "5"> **Evaluation #1: Does Age Play A Factor In COVID-19 Deaths?** </font> 
***

```python
# Drop all necessary columns except the ones that have a direct relation to COVID-19 tests, deaths, and various age groups
covid_age_df = covid_df.drop(['Population', 'Pop Density', 'Gini', 'ICU Beds', 'Income', 'GDP', 'Unemployment', 'Sex Ratio', 
                              'Smoking Rate', 'Flu Deaths', 'Respiratory Deaths', 'Physicians', 'Hospitals', 'Health Spending',
                              'Pollution', 'Med-Large Airports', 'Temperature', 'Urban', 'School Closure Date'], axis = 1)
# Print new table 
covid_age_df
```

```python
plt.figure(figsize = (12, 8))

x = np.array(covid_age_df['Age 0-25'])

y = np.array(covid_age_df['Deaths'])

plt.plot(x, y, 'o', color = 'red')

m, b = np.polyfit(x, y, 1)

plt.plot(x, (m * x) + b, color = 'black')

plt.grid()

plt.title("Age Groups 0-25 vs. Deaths", fontsize = 20)

plt.ylabel("Deaths", fontsize = 18)

plt.xlabel("Ages 0-25", fontsize = 18)
```

```python
covid_age_df['age_group_one'] = covid_age_df['Age 0-25']

# Print Regression Results for Age Group 0-25
regression_results = sm.ols(formula =  'Deaths ~ age_group_one', data = covid_age_df).fit()

# Create first parameter for linear regression model 
first_param = [[x] for x in covid_age_df['Age 0-25']]

# Create second parameter for linear regression model
second_param = [[y] for y in covid_age_df['Deaths']]

# Create a variable to call the linearRegression import above
lin_regression = linear_model.LinearRegression()

# Create variable to store linear regression model average data
lin_reg_model = lin_regression.fit(first_param, second_param) 

# Create coefficient for the above variable
format_one = lin_reg_model.coef_[0]

#Obtain y-intercept of above variable 
format_two = lin_reg_model.intercept_

# Print the data for the fitted model 
print (regression_results.summary())

# Print new line for spacing purposes
print ()

# prints on average how much life expentancy increases every year around the world
print ("Linear Regression Model (Slopt Intercept Form): Y = {0}x + {1}".format(format_one, format_two))

# Print new line for spacing purposes
print ()

# Find the p-value 
slope, intercept, r_value, p_value, stderr = stats.linregress(covid_age_df['Age 0-25'], covid_age_df['Deaths'])

# Print the p-value 
print ("The P-Value is: ", p_value)
```
